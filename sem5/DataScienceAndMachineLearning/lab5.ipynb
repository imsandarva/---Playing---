# Scatter Plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("IRIS.csv")

# Seaborn scatter plot with hue (categorical coloring)
plt.figure(figsize=(12, 6))

sns.scatterplot(data=df, x='sepal_length', y='sepal_width',
                hue='species')

plt.title('Showing length and width of iris flower species in Scatter Plot.', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

# Line plot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("IRIS.csv")

# Seaborn scatter plot with hue (categorical coloring)
plt.figure(figsize=(12, 6))

sns.lineplot(data=df, x='sepal_length', y='sepal_width',
                hue='species')

plt.title('Showing length and width of iris flower species in Line plot', fontsize=14, fontweight='bold')
plt.legend(bbox_to_anchor=(1.02, 1), loc='upper left')

plt.tight_layout()
plt.show()
```

# Histogram

```python

import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("IRIS.csv")

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
sns.histplot(data=df, x='sepal_length', kde=True,
             color='steelblue', edgecolor='white')
plt.title('Iris flower Histogram with KDE', fontweight='bold')

plt.subplot(1, 2, 2)
sns.kdeplot(data=df, x='sepal_length', fill=True,
            color='steelblue', alpha=0.5)
plt.title('Iris Flower KDE Only', fontweight='bold')

plt.tight_layout()
plt.show()
```

# Barplot or Countplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

df = pd.read_csv("IRIS.csv")

# Grouped bar plot with Seaborn
plt.figure(figsize=(12, 6))

sns.barplot(data=df, x='sepal_length', y='sepal_width', hue='species',
            palette='Set2', edgecolor='white')

plt.xlabel('Length of flower', fontsize=12)
plt.ylabel('Width of flower', fontsize=12)
plt.title('Showing length and width of sepal flower from Histogram', fontsize=14, fontweight='bold')
plt.legend(title='Sepal Flower Visualization')
plt.xticks(rotation=15)

plt.show()
```

# Boxplot

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns


df = pd.read_csv("IRIS.csv")

plt.figure(figsize=(12,6))
sns.boxplot(data=df, x="sepal_length", y="sepal_width", hue="species")

plt.xlabel("Length of flower")
plt.ylabel("Width of flower")

plt.title("Boxplot of Iris flower dataset")

plt.show()
```
